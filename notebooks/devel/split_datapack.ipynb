```python
from bayes_tec.datapack import DataPack
import numpy as np
import pylab as plt
```

```python
with DataPack('../../scripts/data/killms_datapack_4.hdf5') as datapack:
    datapack.select(ant='RS*',time=slice(1,120,1))
    phase, axes = datapack.phase
    amp, axes = datapack.amplitude
```

```python
gains = amp * np.exp(1j*phase)
```

```python
g_mu = np.mean(gains,-2)
plt.plot(g_mu[0,0,4,:].real,g_mu[0,0,4,:].imag)
plt.show()
```
