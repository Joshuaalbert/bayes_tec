```python
import numpy as np
from scipy.special import i0e
import pylab as plt
plt.style.use('ggplot')
from IPython.display import HTML

def vonmises(y, tau, freq, sigma):
    phase = -8.448e9*tau/freq
    kappa = 1/sigma**2
    log_prob = kappa * np.cos(y - phase)
   
    norm = np.log(2*np.pi*i0e(kappa)) + kappa
    return log_prob - norm#- np.log(2*np.pi) - kappa - np.log(i0e(kappa))#- np.log(norm)

def analytic(Y,Fmu, Fvar, sigma):
    A = -8.448e9 / freq
    kappa = 1./sigma**2
    norm = np.log(2*np.pi*i0e(kappa)) + kappa
    return kappa * np.cos(A*Fmu - Y) * np.exp(-A**2 * Fvar/2.)  -  norm


tau = 0.05
freq = 140e6
y = -8.448e9*tau/freq

# sample
sd = 0.005
mu = 0.005
Z = np.random.normal(mu,sd,size=1000)

lik = vonmises(y, Z, freq, 0.1)
print(np.mean(lik))
print(analytic(y, mu, sd**2, 0.1))
```
